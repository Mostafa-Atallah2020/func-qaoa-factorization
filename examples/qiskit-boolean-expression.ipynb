```python
from qiskit.circuit.classicalfunction import BooleanExpression
from qiskit import transpile
```

We wish to get a quantum circuit for the expression `p_5*q_3 + p_6*q_2 + p_7*q_1`

```python
# rewrite the expression in a form readable to qiskit

expr = "p_5 & q_3 | p_6 & q_2 | p_7 & q_1"

# create a Boolean expression class instance
expr = BooleanExpression(expr)
```

```python
# generate a quantum circuit from the expression
circuit = expr.synth()

# draw the circuit
circuit.draw("mpl", style="iqx")
```
