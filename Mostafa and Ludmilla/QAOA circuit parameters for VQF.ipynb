```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
max_pow = 5
```

```python
level = 0
file_0 = f"./qaoa_data/qaoa_circ_params_max_power_{max_pow}_transpile_level_{level}.csv"
df_0 = pd.read_csv(file_0)
depth_0 = list(df_0["depth"].values)
n_gates_0 = list(df_0["n_gates"].values)
n_cnot_0 = list(df_0["n_cnot"].values)
n_u3_0 = list(df_0["n_u3"].values)
```

```python
level = 1
file_1 = f"./qaoa_data/qaoa_circ_params_max_power_{max_pow}_transpile_level_{level}.csv"
df_1 = pd.read_csv(file_1)
depth_1 = list(df_1["depth"].values)
n_gates_1 = list(df_1["n_gates"].values)
n_cnot_1 = list(df_1["n_cnot"].values)
n_u3_1 = list(df_1["n_u3"].values)
```

```python
level = 2
file_2 = f"./qaoa_data/qaoa_circ_params_max_power_{max_pow}_transpile_level_{level}.csv"
df_2 = pd.read_csv(file_2)
depth_2 = list(df_2["depth"].values)
n_gates_2 = list(df_2["n_gates"].values)
n_cnot_2 = list(df_2["n_cnot"].values)
n_u3_2 = list(df_2["n_u3"].values)
```

```python
level = 3
file_3 = f"./qaoa_data/qaoa_circ_params_max_power_{max_pow}_transpile_level_{level}.csv"
df_3 = pd.read_csv(file_3)
depth_3 = list(df_3["depth"].values)
n_gates_3 = list(df_3["n_gates"].values)
n_cnot_3 = list(df_3["n_cnot"].values)
n_u3_3 = list(df_3["n_u3"].values)
```

```python
n_qubit = list(df_0["n_qubits"].values)
m = list(df_0["biprime"].values)
```

## Circuit Depth vs Biprimes to be factored

```python
plot_name = f"./qaoa_plots/qaoa_circ_width_max_power_{max_pow}_plot.png"
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(m, depth_0, marker="o", label="Transpilation Level 0")
ax.plot(m, depth_1, marker="o", label="Transpilation Level 1")
ax.plot(m, depth_2, marker="o", label="Transpilation Level 2")
ax.plot(m, depth_3, marker="o", label="Transpilation Level 3")

ax.set_ylabel("Circuit Depth")
ax.set_xlabel("Biprimes to be factored")
ax.set_xscale("log", base=2)

plt.legend()
plt.savefig(plot_name)
plt.show()
```

## Number of Gates vs. Biprimes to be Factored

```python
plot_name = f"./qaoa_plots/qaoa_n_gates_max_power_{max_pow}_plot.png"
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(m, n_gates_0, marker="o", label="Transpilation Level 0")
ax.plot(m, n_gates_1, marker="o", label="Transpilation Level 1")
ax.plot(m, n_gates_2, marker="o", label="Transpilation Level 2")
ax.plot(m, n_gates_3, marker="o", label="Transpilation Level 3")

ax.set_ylabel("Number of Gates")
ax.set_xlabel("Biprimes to be factored")
ax.set_xscale("log", base=2)

plt.legend()
plt.savefig(plot_name)
plt.show()
```

## Number of CNOTS vs Biprimes to be Factored

```python
plot_name = f"./qaoa_plots/qaoa_n_cnot_max_power_{max_pow}_plot.png"
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(m, n_cnot_0, marker="o", label="Transpilation Level 0")
ax.plot(m, n_cnot_1, marker="o", label="Transpilation Level 1")
ax.plot(m, n_cnot_2, marker="o", label="Transpilation Level 2")
ax.plot(m, n_cnot_3, marker="o", label="Transpilation Level 3")

ax.set_ylabel("Number of CNOTS")
ax.set_xlabel("Biprimes to be factored")
ax.set_xscale("log", base=2)

plt.legend()
plt.savefig(plot_name)
plt.show()
```
